```python
import pandas as pd
import sklearn

from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train_path= 'data/X_train.csv'
X_test_path= 'data/X_test.csv'
y_train_path= 'data/y_train.csv'
y_test_path= 'data/1234.csv'
```

```python
X_train = pd.read_csv(X_train_path, encoding='cp949')
X_train = X_train.iloc[:,1:]
X_test = pd.read_csv(X_test_path, encoding='cp949')
X_test_id = X_test.iloc[:,0]
X_test = X_test.iloc[:,1:]
y_train = pd.read_csv(y_train_path, encoding='cp949')
y_train = y_train.iloc[:,-1]
```

```python
X_train
```

```python
X_test
```

```python
X_train.loc[:,['주구매상품','주구매지점']] = \
X_train.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
print(X_train.loc[:,['주구매상품','주구매지점']])
X_test.loc[:,['주구매상품','주구매지점']] = \
X_test.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
print(X_test.loc[:,['주구매상품','주구매지점']])
```

```python
X_train.loc[:,['환불금액']]= X_train.loc[:,['환불금액']].fillna(0)
X_test.loc[:,['환불금액']]= X_test.loc[:,['환불금액']].fillna(0)
print(X_train.loc[:,['환불금액']])
```

```python
X_train.corr()
```

```python
model= LogisticRegression()
model.fit(X_train, y_train)
print("LR score : {}".format(model.score(X_train, y_train)))
```

```python
predict = model.predict_proba(X_test)
predict = pd.DataFrame(predict)
predict = predict.iloc[:,1]
print(predict)
```

```python
answer = pd.concat([X_test_id, predict], axis=1)
answer.to_csv(y_test_path, index= False)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,), solver='adam', activation= 'relu'
                    , learning_rate_init= 0.001, max_iter= 500 )
mlp.fit(X_train, y_train)
print("NN Training ACC : ", mlp.score(X_train, y_train) )
```

```python
predict = mlp.predict_proba(X_test)
predict = pd.DataFrame(predict)
predict = predict.iloc[:,1]
print(predict)
```

```python
rf = RandomForestClassifier(n_estimators= 100, max_depth= 5, random_state= 999)
rf.fit(X_train, y_train)
print("NN Training ACC : ", rf.score(X_train, y_train) )
```

```python
predict = rf.predict_proba(X_test)
predict = pd.DataFrame(predict)
predict = predict.iloc[:,1]
print(predict)
```
